## генераторы

генерирует последовательность

```python
def ones(n):
    while n > 0:
        n -= 1
        yield 1

for o in ones(4): 
    print(o)
```

```python
class A:
    field0 = 0 # public
    _field1 = 1 # protected
    __field2 = 3 # private 
    
    def __init__(self, field0=12):
        self.field0 = field0
    
    @property
    def field2(self):
        return self.__field2
    
    def method0(self):
        return self.field0+self._field1-self.__field2
    
a = A(33)
a.field0
a._field1

a.field2
# a.__field2 # Error!
a._A__field2

a.method0()
```

```python
class Animal:
    
    # конструктор 
    def __init__(self, name, color):
        self.name = name
        self.color = color

class Wolf(Animal):
    def bark(self):
        print("Grr...!")


class Dog(Wolf):
    def bark(self):
        super().bark()
        print("Woof!")
    def __repr__(self):
        return "Dog({})".format(self.name)    
    
Wolf('J','gray').bark()
Dog('B','black').bark()    
```

### ООП декораторы

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self._allowed = False

    def calculate_area(self):
        return self.width * self.height

    @classmethod
    def new_square(cls, side_length):
        return cls(side_length, side_length)

    @staticmethod
    def square(a):
        return a**2

    @property
    def allowed(self):
        return self._allowed

    @allowed.setter
    def allowed(self, value):
        self._allowed = not(value)
```

```python
sq = Rectangle.new_square(5)

print(sq.calculate_area())
# 25
sq.allowed=0 
print(sq.allowed)
# True
print(Rectangle.square(4))
# 16
```
